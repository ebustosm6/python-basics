## Cosas básicas de ordenadores
### ¿Qué es un ordenador?
Un ordenador es una máquina que sirve para hacer todo tipo de cosas y se compone de dos cosas:
- Hardware (cacbles, chips, etc)
- Software (programas)

¿Y cómo funciona todo eso?

Pues los ordenadores siempre tienen un Software (programa) básico que se llama Sistema Operativo (SO) y que sirve para ejecutar otros programas. Los más habituales son Windows y Linux.

El resto de programas los ejecuta el SO. Por ejemplo, el explorador Google Chrome o Microsoft Word.

Entonces podemos pintar la idea de un ordenador como en la siguiente imagen:

![ordenador](resources/ordenador.png)


### ¿Qué componentes hardware tiene?
Las partes del hardware que nos interesa saber son tres:
- Procesador: Es donde se realizan las operaciones (el que ejecuta los programas).
- Memoria: Es donde se guardan los programas y sus datos mientras se ejecutan. Esta es la memoria RAM y se pierde cuando el ordenador se apaga.
- Sistema de archivos: Es donde se guardan datos (en archivos) para que persistan cuando esté apagado. Por ejemplo, el disco duro o un USB o tarjeta de memoria.

## Cosas básicas del sistema operativo
### El sistema operativo (SO)
Es es programa encargado de ejecutar otros programas. Los sistemas operativos están escritos en lenguajes de programación más complejos (Linux en C, por ejemplo).

### ¿Cómo ejecuta el SO otros programas?
Pues utiliza el procesador para mandarle los programas y ejecutarlos. Los procesadores tienen núcleos, y puede ejecutar a la vez tantos programas como núcleos (siempre son pares, normalmente 4 u 8 núcleos).
Esto quiere decir que sólo se pueden ejecutar 4 u 8 programas al mismo tiempo.
A simple vista parece que todo se ejecuta a la vez pero es mentira, lo que ocurre es que el ordenador va cambiando continuamente de programa en el procesador más rápido de lo que los humanos nos damos cuenta.

### ¿Cómo gestiona el SO esto de mandar a ejecutar cosas?
Pues los ordenadores SO tienen unas cosas llamadas procesos, que son las entidades de procesamiento básicas y que tienen dos cosas: código y datos. El código corresponde al programa (o parte del programa) que se quiere ejecutar y los datos a valores como texto o números. Entonces un programa se ejecuta "dentro" de un proceso.
Los procesos se pueden ver en el administrador de tareas (en windows). Cada proceso tiene un id llamado PID.

![administrador de tareas](./resources/administrador_tareas.png)

Esto implica dos cosas MUY IMPORTANTES:
- Cada proceso tiene su memoria individual.
- Los datos o partes de código no se comparten entre procesos.

### ¿Y para programar, de qué nos sirve?
Pues porque el SO tiene una serie de funciones que podemos llamar desde nuestros programas, como las Interrupciones por ejemplo.

## Cosas básicas de un programa
### ¿Qué es un programa?
Un programa es un conjunto de instrucciones (líneas de código) que hacen una tarea. Estas líneas de código se ejecutan para que el ordenador haga una tarea. Para ello, el programa puede utilizar datos (leídos desde un archivo, que están en la memoria, etc).

### Organización de un programa
Los programas se escriben en un lenguaje de programación y se organizan en carpetas y archivos con una estructura lógica. La idea es separar las partes del progrma de forma que tenga sentido. Por ejemplo, meter los datos en una carpeta de datos, el código en una de código, y luego dentro de esta separar las partes de código de cada cosa para que esté organizado.

### ¿Cómo se ejecuta un programa escrito en un lenguaje diferente al del SO?
Los programas pueden escribirse en muchos lenguajes de programación que son diferentes al del SO.
Para ejecutarlos se puede hacer de dos maneras:
- Con un intérprete (proceso que interpreta el código en un lenguaje y lo traduce al del SO). Como Python en este caso.
- Compilando (se traducen los archivos a otro lenguaje y se crea un archivo ejecutable que entiende un proceso ejecutor o el SO). Por ejemplo C o Java.

## Cosas básicas de los lenguajes de programación
### ¿Qué son?
Son lenguajes, son como los idiomas, que sirven para decirle al SO las operaciones que tiene que hacer. Casi todos los lenguajes pueden hacer las mismas operaciones básicas:
- guardar un valor en un espacio de memoria (llamado variable)
- hacer operaciones con dispositivos o archivos (llamadas operaciones de Entrada/Salida o E/S).
- lanzar Interrupciones (errores que saltan en el prorama).

### ¿Qué tipos hay?
Principalmente dos:
- Compilados: se escribe el código y luego se traduce a un lenguaje máquina para que lo entienda el SO u otro programa. C y Java son compilados, C se traduce a código que entiende el SO y Java se traduce a código que entiende la máquina virtual de java (es el ejecutor de java).
- Interpretados: un proceso intérprete va leyendo el programa y ejecutándolo. Python es un lenguaje interpretado.

### ¿Cuáles son las palabras que usa cada lenguaje?
Las palabras de cada lenguaje (como los idiomas) son propias del lenguaje pero las ideas (o conceptos) son los mismos. Estas son las que hay que saber para escribir programas.

Por ejemplo, las siguientes son comunes a muchos lenguajes y en todos sirven para lo mismo:
- if: evaluar una condición, si pasa eso, se ejecuta lo siguiente.
- else: el caso de que un "if" no se cumpla.
- for: para iterar (recorrer) una lista de cosas de una en una.
- while: para iterar todo el rato hasta que una condición se cumpla o se deje de cumplir.
- try: para intentar ejecutar un trozo de código que puede fallar

U otras que cambian según el lenguaje:
- except (python) y catch (java): para lo que se ejecuta si lo del "try" ha dado un fallo.
